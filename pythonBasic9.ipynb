```python
for i in range(100):
    flag,img=vid.read()
    if flag:
        cv2.imshow('Preview',img)
        cv2.waitKey(1)
cv2.destroyAllWindows()
cv2.waitKey()
```

```python
while True:
    flag,img=vid.read()
    if flag:
        cv2.imshow('Preview',img)
        key=cv2.waitKey(1)
        if key == ord('x'):
            break
vid.release()
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
import cv2
from time import sleep
while True:
    flag,img=vid.read()
    if flag:
        img_gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)



        cv2.imshow('Preview',img_gray)
        Key=cv2.waitKey(1)
        if Key==ord('q'):
            break
    else:
        break
    sleep(0.1)
vid.release()
cv2.destroyAllWindows()
```
